```python
#importing necessary libraries and packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('darkgrid')

#setting random number for die rolling
rolls = [random.randrange(1, 7) for i in range(1000)]
values, frequencies = np.unique(rolls, return_counts = True)


title = f'Rolling a Six-Sided Die {len(rolls):,} Times'
axes = sns.barplot(x = values, y= frequencies, palette ='bright')
plt.title(title)
plt.xlabel('Die Number')
plt.ylabel('Frequency')
plt.ylim(0, top=max(frequencies)* 1.10)
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(rolls):.2%}'
    axes.text(text_x, text_y, text, fontsize=12,
               ha='center', va = 'bottom')   
```

```python
#modifying the code for coin tossing
roll = [random.randrange(1,3) for x in range(2000)]
flip, freq = np.unique(roll, return_counts=True)
coin_title = f'Flipping of Golden Coin {len(roll):,} times'
axes = sns.barplot(x=flip, y=freq, palette= 'bright')
sns.set_style("darkgrid")
axes.set_title(coin_title, fontsize=15, color = 'green')
axes.set_ylabel('Frequency of the flipping', fontsize = 15)
axes.set_xlabel('The tossed side', fontsize = 15)
axes.set_ylim(top = max(freq)*1.10)
ct = [1,2]
ct_lb = ['Head','Tail']
axes.set_xticks(ct, minor = True)
axes.set_xticklabels(ct_lb,{'fontsize': 15,'verticalalignment': 'top','horizontalalignment': 'center'})
for bar, frequency in zip(axes.patches, freq):
    text_x = bar.get_x() + bar.get_width()/2.0
    text_y = bar.get_height()
    text = f'{frequency:,} \n {frequency/len(roll):.0%}'
    axes.text(text_x, text_y, text, fontsize=12,
               ha='center', va = 'bottom')
plt.show()
```

```python
rolls = [sum([random.randint(1,6), random.randint(1,6)]) for i in range(3600)]
roll, frequency = np.unique(rolls, return_counts = True)
sns.set_style('darkgrid')
axes = sns.barplot(x=roll, y=frequency, palette = 'bright')
title = f'Rolling of two dice ({len(rolls):,}) times'
plt.title(title, fontsize = 13)
plt.xlabel('Sum of the two Dice', fontsize=13)
plt.ylabel('Frequency of the Occurence', fontsize=13)
plt.ylim(top = max(frequency)*1.10)
for bar, freq in zip(axes.patches, frequency):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{freq:,} \n {freq/len(rolls):.2%}'
    axes.text(text_x,text_y,text,
             fontsize=12, ha = 'center', va = 'bottom')
plt.show()
```

```python
rolls = [sum([random.randint(1,6), random.randint(1,6)]) for i in range(3600)]
roll, frequency = np.unique(rolls, return_counts = True)
sns.set_style('darkgrid')
axes = sns.barplot(x=frequency, y=roll, palette = 'bright', orient = "h")
title = f'Rolling of two dice ({len(rolls):,}) times'
plt.title(title, fontsize = 13)
plt.ylabel('Sum of the two Dice', fontsize=13)
plt.xlabel('Frequency of the Occurence', fontsize=13)
plt.xlim(right = max(frequency)*1.10)
for bar, freq in zip(axes.patches, frequency):
    text_x = bar.get_x() + bar.get_width() + 2.0
    text_y = bar.get_y() + bar.get_height() / 2.0
    text = f'{freq:,} \n {freq/len(rolls):.1%}'
    axes.text(text_x,text_y,text,
             fontsize=10, ha = 'left', va = 'center', )
plt.show()
```
